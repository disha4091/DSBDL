```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movies_metadata.csv',low_memory=False)
df.head()
```

```python
df = df.dropna()
```

### Subsets

```python
df.original_language.unique()
```

```python
english_sub = df[df.original_language == 'en']
english_sub
```

```python
french_sub = df[df.original_language == 'fr']
french_sub
```

### Merge

```python
merged_data = pd.merge(english_sub,french_sub, how="outer")
merged_data.head()
```

### Sort

```python
df.sort_values('vote_count')
```

### Transpose

```python
df.transpose()
```

### Melting

```python
df_melted = pd.melt(df, id_vars=['original_language'], value_vars=['revenue', 'budget'])
```

```python
df_melted
```

```python
df_pivot = pd.pivot_table(df, values='vote_average', index='original_language', aggfunc="mean")
```
